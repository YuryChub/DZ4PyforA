Изучим связь между общей площадью дома и его стоимостью

```python
plt.figure(figsize=(15,8))
sns.scatterplot(y=df["sqft_living"], x=df["price"]/1000)
plt.ylabel("sqft_living")
plt.xlabel("price (*1000)")
```

Вывод: Прослеживается практически прямая линейная зависимость между площадью дома и его стоимостью.

Изучим связь между этажностью дома и его стоимостью

```python
plt.figure(figsize=(15,8))
# sns.boxplot(y=df['price']/1000, x=df['floors'], showfliers=False)
sns.boxplot(y=df['price']/1000, x=df['floors'])
# sns.boxplot(data=[data1, data2], showfliers=False)
plt.ylabel("price (x 1000)")
plt.xlabel("floors")
```

Вывод: Прослеживается нелинейная зависимость цены дома от его этажности - самые дорогие дома 2,5 этажа

Изучим связь между состоянием дома и его стоимостью

```python
plt.figure(figsize=(15,8))
# sns.boxplot(y=df['price']/1000, x=df['floors'], showfliers=False)
sns.boxplot(y=df['price']/1000, x=df['condition'])
# sns.boxplot(data=[data1, data2], showfliers=False)
plt.ylabel("price (x 1000)")
plt.xlabel("condition")
```

Вывод. Прослеживается прямая зависимось между состоянием дома и его ценой. Однако, для среднего уровня состояниния дома и выше присутсвузначительное количество выбросов, что вероятно свидетедьствует о других факторах, влияющих на стоимость.

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');
```

```python
df.drop(columns=['id'], inplace=True)
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix

plt.figure(figsize=(10, 8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```

Вывод: Прослеживается сильная линенйная взаимосвязь (0,7 и более) между ценой дома и жилой площадью и качеством дома (grade), средняя зависимость (0,5 - 0,6) - площадью дома выше земли, жилой пплощадью домов соседей и количеством ванных комнат. Слабая связь - с видом из окон, общей площадью, количествои этажей и спален.

```python
min_long = df['long'].min()
max_long = df['long'].max()
min_lat = df['lat'].min()
max_lat = df['lat'].max()
print (min_lat, min_long)
print (min_lat, max_long)
print (max_lat, max_long)
print (max_lat, min_long)
```
